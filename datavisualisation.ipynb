### 3D Plot of Iris Dataset Using Axes3D and Matplotlib

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets

iris=datasets.load_iris()
x=iris.data
y=iris.target
labels=['Setosa', 'Versicolor', 'Virginica']
nums=[0,1,2]
colors=['red','blue','green']
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(x[:,0],x[:,1],x[:,2],c=y)
for num, label in zip(nums,labels):
    ax.text3D(x[y == num, 0].mean(),
              x[y == num, 1].mean() ,
              x[y == num, 2].mean(), label,
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))

ax.set_title("3D Plot of Iris Dataset")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
plt.show()
```

###                                                 3D Plot of Iris Dataset using Seaborn library


```python
import seaborn as sns

iris=sns.load_dataset('iris') 

fig = plt.figure()
ax=plt.axes(projection='3d')
ax.scatter(iris['petal_length'],iris['petal_width'],iris['sepal_length'],c=iris['sepal_length'])
ax.set_xlabel('Petal Length')
ax.set_ylabel('Petal Width')
ax.set_zlabel('Sepal Length')
ax.set_title("Seaborn plot of Iris Dataset ")
```
