```python
import pandas as pd
```

```python
%%time
df1=pd.read_csv('household_power_consumption.txt',sep=';')    
```

```python
%%time 
df1.describe()
```

```python
df1.info()
```

```python
%%time
df1.dropna(axis=0,inplace=True)
```

```python
df1['Sub_metering_1']=df1['Sub_metering_1'].astype(float)
df1['Sub_metering_2']=df1['Sub_metering_2'].astype(float)
df1['Global_active_power']=df1['Global_active_power'].astype(float)
```

```python
%%time
GroupbyDate=df1.groupby('Date')
GroupbyDate.sum()
```

```python
df1['Date']=pd.to_datetime(df1['Date'])
df1['Year']=df1['Date'].dt.year
df1['Month']=df1['Date'].dt.month
```

```python
%%time 
Yearwisegroup=df1.groupby(['Year'])
```

```python
%%time
Yearwisegroup.agg({'Sub_metering_1':['sum'],'Sub_metering_2':['sum'],'Sub_metering_3':['sum'],'Global_active_power':['mean']})
```

```python
%%time 
Monthwisegrp=df1.groupby(['Year','Month'])
Monthwisegrp.agg({'Sub_metering_1':['sum'],'Sub_metering_2':['sum'],'Sub_metering_3':['sum'],'Global_active_power':['mean']})
```
