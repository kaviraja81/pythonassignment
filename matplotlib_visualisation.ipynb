```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import optimize
```

## Fitting the periodic function

```python
tempmax=[39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
months=[i for i in range(1,13)]
tempmin=[21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]

def temp_fit(month,avg,ampl,offset):
    return avg+ampl*np.cos((month+offset)*2*np.pi/month.max())

optmin,covmin=optimize.curve_fit(temp_fit,months,tempmin,[20, 10, 0])
optmax,covmax=optimize.curve_fit(temp_fit,months,tempmax,[20, 10, 0])
print(optmin,covmin)
```

## Plot the Temperature Vs Months as well as plotting the fitting line

```python
days=np.linspace(0,12)
fig,ax=plt.subplots()
ax.scatter(months,tempmin,label="Minimum Temp")
ax.plot(days,temp_fit(days,*optmin))
ax.scatter(months,tempmax,label="Maximum Temp")
ax.plot(days,temp_fit(days,*optmax))
ax.set_xlabel('Months')
ax.set_ylabel('Temperature(C)')
ax.legend(loc='upper right')
ax.set_title("Temperature Vs Month")
plt.show()
```

## Taking a titanic dataset as input , plot a pie  chart and a scatter plot
### Plot a pie chart presenting the male/female propotion

```python
import pandas as pd
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
l1=list(titanic['sex'].value_counts())
fig,ax=plt.subplots()
labels=['Male','Female']
ax.pie(l1,labels=labels,shadow=True,autopct='%1.1f%%')
plt.show()
```

## Plot a scatter plot with Fare paid Vs Age

```python
import seaborn as sns
ax=sns.scatterplot(x = 'age', y = 'fare', data = titanic, hue = 'sex')
ax.set(xlabel='Age', ylabel='Fare',title='Fare Vs Age')
```
